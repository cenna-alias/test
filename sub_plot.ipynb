```python
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.plot([1,2,3], [4,5,6], marker='o')
plt.title("Plot 1")

plt.subplot(2, 2, 2)
plt.plot([1,2,3], [6,5,4], marker='s')
plt.title("Plot 2")

plt.subplot(2, 2, 3)
plt.plot([1,2,3], [2,3,1], marker='^')
plt.title("Plot 3")

plt.subplot(2, 2, 4)
plt.plot([1,2,3], [3,1,2], marker='*')
plt.title("Plot 4")

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot([0, 1, 2, 3], [3, 8, 1, 10], marker='o')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot([0, 1, 2, 3], [10, 20, 30, 40], marker='o')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot([0, 1, 2, 3], [40, 30, 20, 10], marker='o')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot([0, 1, 2, 3], [10, 1, 8, 3], marker='o')
plt.grid(True)

plt.tight_layout()
plt.show()
```
