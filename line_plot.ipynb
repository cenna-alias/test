```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]
sales = [10,20,30,35,15,45,25]

plt.plot(days, sales, linestyle=':', marker='.', label='Sales', color='purple')

plt.xlabel("Days")
plt.ylabel("Sale in each day")
plt.title("Basic line plot for sales analysis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

x = np.arange(1, 11)

y1 = x**2
y2 = 2*x

plt.plot(x, y1, label='square', linestyle=':', marker='o', color='red')
plt.plot(x, y2, label='double', linestyle='-', marker='s', color='blue')

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Plot of square and double")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.arange(1,21)
y = x*2
plt.scatter(x, y, s=10, label='double', color='green')

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,16)
y = np.random.randint(1000, 9999, 15)
plt.scatter(x, y, s=50, facecolors='red', label='random values', edgecolors='blue')

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['A','B','C']
values = (10,20,15)
plt.bar(categories,values)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Bar Chart")
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values,color='green', edgecolor='blue', linewidth=2, label='Values')

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.legend()
plt.show()
```

```python
categories = ['A','B','C','D','E']
values = (50,20,30,35,45)
select = ['teal','coral','gold','olive','gray']
plt.bar(categories, values, color=select, edgecolor='yellow', linewidth=5)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Bar Chart")
plt.show()
```

```python
weight = (50,70,85,90,45,65,76,87,44,60)
height = (150,176,185,190,145,165,176,187,144,160)
plt.scatter(weight, height, s=50, facecolors='yellow', edgecolors='green', label='Students')
plt.xlabel("Weight(kg)")
plt.ylabel("Height(cm)")
plt.legend()
plt.title("Height and Weight analysis")
plt.show()
```

```python
product_list = ['Bag','Shoes','Salwar','Perfumes','Watch']
sales_amount = (999,1899,2900,599,1999)
select = ['orange','blue','black','red','purple']
plt.bar(product_list, sales_amount, color=select, edgecolor='yellow', linewidth=5, label='Sales')
plt.xlabel("Product_list")
plt.ylabel("Sales_amount")
plt.legend()
plt.title("Product Analysis")
plt.show()
```
