```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]
sales = [10,20,30,35,15,45,25]

plt.plot(days,sales,linestyle=':', marker='.', label='Sales',color='green')

plt.xlabel("Days")
plt.ylabel("Sale in each day")
plt.title("Basic line plot for sales analysis")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)

y1 = x**2
y2 = 2*x

plt.plot(x, y1, label='square', linestyle=':', marker='o', color='red')
plt.plot(x, y2, label='double', linestyle='-', marker='s', color='blue')

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Plot of square and double")
plt.legend()
plt.grid(True)
plt.show()
```
