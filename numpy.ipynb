## 1️⃣ Why NumPy? (Speed vs Python Lists)
Python lists:
- Store elements in different memory locations
- Can store mixed data types
- Each element is a Python object
- Operations run through Python interpreter (slow for large data)

NumPy:
- Stores data in contiguous memory
- Same datatype
- Implemented in C
- Supports vectorized operations (very fast)


```python
# Python list addition (slow approach)
a = [1,2,3]
b = [4,5,6]
result = []
for i in range(len(a)):
    result.append(a[i] + b[i])
print(result)
```

```python
# NumPy vectorized addition (fast approach)
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Array a:",a)
print("Array b:",b)
print(a + b)
```

## 4️⃣ Array Attributes
ndim → Returns the number of dimensions (axes) of the array.

shape → Returns the dimensions of the array as a tuple (rows, columns, …).

size → Returns the total number of elements in the array.

dtype → Returns the data type of the array elements.

itemsize → Returns the memory size (in bytes) of each array element.


### 1 Dimensional array

```python
a1=np.array(['a','b','c','d'])

print("Array:",a1)
print("Dimension:",a1.ndim)
print("Shape:",a1.shape)
print("Total elements:",a1.size)
print("Data type:",a1.dtype)
print("Memory size:",a1.itemsize)
```

###  Dimensional array

```python
a = np.array([1,2,3])
print('number of dimension:', a.ndim)
print('shape:', a.shape)
print('size:', a.size)
print('dtype:', a.dtype)
print('itemsize:', a.itemsize)
```

```python
a = np.array([[1,2,3],[4,5,6]])
print('number of dimension:', a.ndim)
print('shape:', a.shape)
print('size:', a.size)
print('dtype:', a.dtype)
print('itemsize:', a.itemsize)
```

## 5️⃣ Basic Element-wise Operations


```python
a = np.array([10,20,30])
b = np.array([1,2,3])
print('Addition:', a + b)
print('Subtraction:', a - b)
print('Multiplication:', a * b)
print('Division:', a / b)
print('Floor Division:', a // b)
print('Scalar Addition:', a + 5)
print('Scalar Multiplication:', a * 5)
```
