```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df[['Age','Annual Income (k$)','Spending Score (1-100)']])
plt.show()
```

```python
X = df[['Annual Income (k$)','Spending Score (1-100)']]

X.head()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

```python
wc = []

for i in range(2,11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X_scaled)

    wc.append(kmeans.inertia_)

plt.figure(figsize=(8,5))

plt.plot(range(2,11),wc,marker='o')

plt.xlabel('Number of Clusters')
plt.ylabel('WC')
plt.title('Elbow Method')

plt.show()
```

```python
kmeans4 = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

clusters = kmeans4.fit_predict(X_scaled)

df['Cluster'] = clusters

df.head()
```

```python
df['Cluster'].value_counts()
```

```python

centers = scaler.inverse_transform(kmeans4.cluster_centers_)

centers_df = pd.DataFrame(
    centers,
    columns=['Annual Income','Spending Score']
)

centers_df
```

```python

plt.figure(figsize=(10,6))

plt.scatter(
    df['Annual Income (k$)'],
    df['Spending Score (1-100)'],
    c=df['Cluster']
)

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments Using K-Means 4')

plt.show()
```
