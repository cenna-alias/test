```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import numpy as np

# Days (X-axis)
days = np.arange(7)  # Days 0-6

# Sample stock prices
apple = [150, 152, 153, 151, 155, 157, 156]
google = [2800, 2820, 2810, 2830, 2840, 2850, 2860]
amazon = [3400, 3420, 3410, 3430, 3440, 3450, 3460]
microsoft = [299, 301, 300, 302, 303, 305, 304]

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot Apple
axs[0, 0].plot(days, apple, marker='o', color='red')
axs[0, 0].set_title("Apple Stock Price")
axs[0, 0].grid(True)

# Plot Google
axs[0, 1].plot(days, google, marker='s', color='blue')
axs[0, 1].set_title("Google Stock Price")
axs[0, 1].grid(True)

# Plot Amazon
axs[1, 0].plot(days, amazon, marker='^', color='green')
axs[1, 0].set_title("Amazon Stock Price")
axs[1, 0].grid(True)

# Plot Microsoft
axs[1, 1].plot(days, microsoft, marker='d', color='purple')
axs[1, 1].set_title("Microsoft Stock Price")
axs[1, 1].grid(True)

# Add overall X and Y labels
for ax in axs[1, :]:
    ax.set_xlabel("Day")

for ax in axs[:, 0]:
    ax.set_ylabel("Price ($)")

plt.tight_layout()
plt.show()
```
